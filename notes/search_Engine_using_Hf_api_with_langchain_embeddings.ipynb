```python
def fun(query, docs):
  query_embed = hf_embedings.embed_query(query)
  docum_embed = hf_embedings.embed_documents(docs)
  cosine_sim = cosine_similarity([query_embed], docum_embed)
  cosine_sim = cosine_similarity([query_embed], docum_embed)
  print(docs[np.argmax(cosine_sim[0])])
```
